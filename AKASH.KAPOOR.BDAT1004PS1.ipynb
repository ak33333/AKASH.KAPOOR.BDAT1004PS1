using System;
					
public class Program
{
	public static double triangleArea(int a, int b, int c)
	{
		double s = (a + b + c) / 2;
		double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
		return area;
	}

	public static void Main()
	{
		Console.WriteLine(">>> triangleArea(2,2,2) is " + triangleArea(2, 2, 2));


	}
}

using System;
					
public class Program
{
	public static bool inside(double x, double y, double x1, double y1, double x2, double y2){
		if ((x >= x1 && x <= x2) && (y >= y1 && y <= y2)){
			return true;
		} else {
			return false;
		}
	}
	
	
	public static void Main()
	{
		Console.WriteLine(">>> inside(1,1,0,0,2,3)   " + inside(1,1,0,0,2,3));
		Console.WriteLine(">>> inside(-1,-1,0,0,2,3) " + inside(-1,-1,0,0,2,3));
		
	}
}

```python
def pig(word):
    vowels = ["a","e","i","o","u"]
    if word[0].lower() in vowels:
        return (word + "way").lower()
    else:
        return (word[1:] + word[0] + "ay").lower()
pig("Enter")
```

# Question 8

```python
def curconv(currency, amount):
    lines = open("currencies.txt").readlines()
    currencies = {}
    for line in lines:
        data = line.split(" ")
        currencies[data[0]] = float(data[1])
    #print(currencies)
    return currencies[currency] * amount
print(">>> curconv('EUR', 100)" , curconv('EUR', 100))
```

```python
letters = "abcdefghijklmnopqrstuvwxyz"
def frequencies(words):
    res = [0] * 26
    for char in words:
        pos = letters.find(char)
        if pos >= 0:
            res[pos] = res[pos] + 1
    return res
frequencies('The quick red fox got bored and went home.')
```
